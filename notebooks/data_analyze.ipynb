```python
# %%
import json
from pathlib import Path
import pandas as pd

DATA_DIR = Path("../data/datasets/merged_splits")

def load_jsonl(path: Path) -> pd.DataFrame:
    records = []
    with path.open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            records.append(json.loads(line))
    return pd.DataFrame(records)

train_df = load_jsonl(DATA_DIR / "train.jsonl")
val_df   = load_jsonl(DATA_DIR / "val.jsonl")
test_df  = load_jsonl(DATA_DIR / "test.jsonl")

for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
    print(name, df.shape)


# %%
def show_severity(df, name):
    print(f"\n=== {name.upper()} — severity counts ===")
    print(df["severity"].value_counts(dropna=False))
    print(f"\n=== {name.upper()} — severity proportions ===")
    print(df["severity"].value_counts(normalize=True, dropna=False).round(3))

for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
    show_severity(df, name)


# %%
def show_topic(df, name, top_n=20):
    print(f"\n=== {name.upper()} — topic counts (top {top_n}) ===")
    print(df["topic"].value_counts().head(top_n))
    print(f"\n=== {name.upper()} — topic proportions (top {top_n}) ===")
    print(df["topic"].value_counts(normalize=True).head(top_n).round(3))

for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
    show_topic(df, name)


# %%
for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
    print(f"\n=== {name.upper()} — source counts ===")
    print(df["source"].value_counts())
    print(f"\n=== {name.upper()} — source proportions ===")
    print(df["source"].value_counts(normalize=True).round(3))
```

```python
# %%
# Summary table: severity distribution by split (counts + percentages)

def severity_summary_table(train_df, val_df, test_df):
    rows = []
    for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
        n = len(df)
        counts = df["severity"].value_counts(dropna=False)
        props = df["severity"].value_counts(normalize=True, dropna=False)

        for sev in sorted(counts.index):
            rows.append(
                {
                    "split": name,
                    "severity": sev,
                    "count": counts[sev],
                    "percent": round(props[sev] * 100, 1),
                    "n_split": n,
                }
            )
    return pd.DataFrame(rows)

sev_summary_df = severity_summary_table(train_df, val_df, test_df)
sev_summary_df
```

```python
# %%
# Summary table: top topics per split

def topic_summary_table(train_df, val_df, test_df, top_n=15):
    rows = []
    for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
        counts = df["topic"].value_counts().head(top_n)
        props = df["topic"].value_counts(normalize=True).head(top_n)

        for topic in counts.index:
            rows.append(
                {
                    "split": name,
                    "topic": topic,
                    "count": counts[topic],
                    "percent": round(props[topic] * 100, 1),
                }
            )
    return pd.DataFrame(rows)

topic_summary_df = topic_summary_table(train_df, val_df, test_df, top_n=15)
topic_summary_df
```

```python
# %%
# Source distribution by split

def source_summary_table(train_df, val_df, test_df):
    rows = []
    for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
        n = len(df)
        counts = df["source"].value_counts()
        props = df["source"].value_counts(normalize=True)

        for src in counts.index:
            rows.append(
                {
                    "split": name,
                    "source": src,
                    "count": counts[src],
                    "percent": round(props[src] * 100, 1),
                    "n_split": n,
                }
            )
    return pd.DataFrame(rows)

source_summary_df = source_summary_table(train_df, val_df, test_df)
source_summary_df
```

```python
df# %%
def length_stats(df, name):
    texts = df["output"]
    char_lens = texts.str.len()
    token_lens = texts.str.split().apply(len)

    return {
        "split": name,
        "n": len(df),
        "mean_chars": char_lens.mean(),
        "median_chars": char_lens.median(),
        "mean_tokens": token_lens.mean(),
        "median_tokens": token_lens.median(),
    }

length_rows = []
for name, df in [("train", train_df), ("val", val_df), ("test", test_df)]:
    length_rows.append(length_stats(df, name))

length_stats_df = pd.DataFrame(length_rows).round(1)
length_stats_df
```
